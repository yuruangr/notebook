```python
file = "dictionary.csv"
countries_df = pd.read_csv(file, decimal=",")
countries_df.head(3)
```

```python
file = "summer.csv"
summer_df = pd.read_csv(file, decimal=",")
summer_df.head(3)
```

```python
summer_corrected_df = summer_df.rename(columns ={'Country':'Code'})
summer_corrected_df = summer_corrected_df.set_index('Code')
summer_corrected_df
```

```python
summer_merged_df = summer_corrected_df.merge(countries_df, on='Code', how='outer')
summer_merged_df = summer_merged_df.set_index('Code')
summer_merged_df
```

```python
summer_countrys_df = summer_merged_df.groupby('Country').count()
summer_medals_only = summer_countrys_df[summer_countrys_df['Medal'] != 0]

Venezuela_medals_total = summer_medals_only.loc['Venezuela','Medal']
Venezuela_medals_total

summer_gold_medals_only = summer_merged_df[summer_merged_df['Medal'] == 'Gold'].groupby('Country').count()
summer_silver_medals_only = summer_merged_df[summer_merged_df['Medal'] == 'Silver'].groupby('Country').count()
summer_bronze_medals_only = summer_merged_df[summer_merged_df['Medal'] == 'Bronze'].groupby('Country').count()

country = "Australia"
summer_bronze_medals_only
```

```python
Medals_dict = {'Country':[],'Medals':[{'Gold': [], 'Silver': [], 'Bronze': []}]}

for index in range(0,len(summer_medals_only.index.values)):
    country = summer_medals_only.index.values[index]
    Medals_dict['Country'].append(summer_medals_only.index.values[index])
    Medals_dict['Medals'][0]['Gold'].append(0)
    Medals_dict['Medals'][0]['Silver'].append(0)
    Medals_dict['Medals'][0]['Bronze'].append(0)

len(Medals_dict['Country'])
```

```python
for index_total in range(0,len(summer_medals_only.index.values)):
    country_all = summer_medals_only.index.values[index_total]
    for index_gold in range(0, len(summer_gold_medals_only.index.values)):
        country_gold = summer_gold_medals_only.index.values[index_gold]
        if country_gold == country_all:
            Medals_dict['Medals'][0]['Gold'][index_total] = summer_gold_medals_only.loc[country_gold,'Medal']
    for index_silver in range(0, len(summer_silver_medals_only.index.values)):
        country_silver = summer_silver_medals_only.index.values[index_silver]
        if country_silver == country_all:
            Medals_dict['Medals'][0]['Silver'][index_total] = summer_silver_medals_only.loc[country_silver,'Medal']
    for index_bronze in range(0, len(summer_bronze_medals_only.index.values)):
        country_bronze = summer_bronze_medals_only.index.values[index_bronze]
        if country_bronze == country_all:
            Medals_dict['Medals'][0]['Bronze'][index_total] = summer_bronze_medals_only.loc[country_bronze,'Medal']

Country = [0, 0, 0]
for index in range(0,len(summer_medals_only.index.values)):
    if summer_medals_only.index.values[index] == 'Belgium':
        Country[0] = Medals_dict['Medals'][0]['Gold'][index]
        Country[1] = Medals_dict['Medals'][0]['Silver'][index]
        Country[2] = Medals_dict['Medals'][0]['Bronze'][index]
      
Country
```
